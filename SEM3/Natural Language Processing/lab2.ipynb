```python
import nltk
import re
import matplotlib.pyplot as plt
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams
text = "The horology of mechanical watches represents a captivating blend of artistry, precision engineering, and timeless tradition. At its core, mechanical horology encompasses the intricate mechanisms that power these watches, often comprised of hundreds of precisely crafted components working in harmony. The first of its kind came to form in the 1940. From the mesmerizing dance of gears and springs in a manual-wind movement to the seamless oscillation of a balance wheel in an automatic movement, each element plays a crucial role in measuring time with remarkable accuracy. Since 1940 beyond mere functionality, mechanical watches also serve as expressions of craftsmanship and heritage, with skilled artisans meticulously assembling and finishing each piece by hand. In the 2000, the allure of mechanical horology lies not only in its ability to keep time but also in its ability to evoke a sense of nostalgia and appreciation for the artistry of a bygone era."
```

a. Count the number of unique tokens in the text.

```python
nltk.download('punkt')
tokens = word_tokenize(text)
unique_tokens_count = len(set(tokens))
print("Number of unique tokens:", unique_tokens_count)
```

b. Count the number of punctuation characters and remove punctuations.

```python
punctuations = re.findall(r'[^\w\s]', text)
punctuation_count = len(punctuations)
text_without_punctuation = re.sub(r'[^\w\s]', '', text)
print("Number of punctuation characters:", punctuation_count)
print("Text without punctuation:", text_without_punctuation)
```

c. Plot the distribution of stopwords used in the text using barplot.

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stopwords_in_text = [word for word in tokens if word.lower() in stop_words]
stopword_distribution = Counter(stopwords_in_text)
plt.bar(stopword_distribution.keys(), stopword_distribution.values(), color='orchid')
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.title('Distribution of Stopwords')
plt.xticks(rotation=45)
plt.show()
```

d. Remove the stopwords from the text.

```python
filtered_text = [word for word in tokens if word.lower() not in stop_words]
print(filtered_text)
```

e. Plot the distribution of each POS Tag using a barplot.

```python
nltk.download('averaged_perceptron_tagger')
pos_tags = pos_tag(tokens)
pos_distribution = Counter(tag for word, tag in pos_tags)
plt.bar(pos_distribution.keys(), pos_distribution.values(), color='orchid')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags')
plt.xticks(rotation=45)
plt.show()
```

f. Determine the number of unique lemma available in the text.

```python
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in tokens]
unique_lemmas_count = len(set(lemmas))
print("Number of unique lemmas:", unique_lemmas_count)
```

g. Plot the frequency distribution of words in the text. Display only the top 10 (most occurring) tokens in the chart.

```python
word_frequency = Counter(filtered_text)
top_10_tokens = word_frequency.most_common(10)
plt.bar(*zip(*top_10_tokens), color='orchid')
plt.xlabel('Tokens')
plt.ylabel('Frequency')
plt.title('Top 10 Tokens')
plt.xticks(rotation=45)
plt.show()
```

h. Find the number of unique bigrams, trigrams and quadgrams (n=4) in the corpus

```python
bigrams = ngrams(filtered_text, 2)
trigrams = ngrams(filtered_text, 3)
quadgrams = ngrams(filtered_text, 4)
unique_bigrams_count = len(set(bigrams))
unique_trigrams_count = len(set(trigrams))
unique_quadgrams_count = len(set(quadgrams))
print("Number of unique bigrams:", unique_bigrams_count)
print("Number of unique trigrams:", unique_trigrams_count)
print("Number of unique quadgrams:", unique_quadgrams_count)
```

i. Find all dates and convert them to the DD-MM-YYYY format.

```python
from datetime import datetime
dates = re.findall(r'\b(?:0?[1-9]|[12][0-9]|3[01])-(?:January|February|March|April|May|June|July|August|September|October|November|December)-\d{4}\b', text)
converted_dates = []
for date in dates:
    dt_obj = datetime.strptime(date, '%d-%B-%Y')
    converted_dates.append(dt_obj.strftime('%d-%m-%Y'))
print("Dates found:", converted_dates)
```

j. Plot a distribution of the different values of year occurring in the text.

```python
years = re.findall(r'\b\d{4}\b', text)
year_distribution = Counter(years)
plt.bar(year_distribution.keys(), year_distribution.values(), color='orchid')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Distribution of Years')
plt.xticks(rotation=45)
plt.show()
```

k. Determine whether the text contains any phone numbers in it.

```python
phone_numbers = re.findall(r'\b(?:\+?(\d{1,2}))?[-. ]?(\d{3})[-. ]?(\d{3})[-. ]?(\d{4})\b', text)
valid_phone_numbers = []
for number in phone_numbers:
    formatted_number = '-'.join(number)
    valid_phone_numbers.append(formatted_number)
if valid_phone_numbers:
    print("Valid phone numbers found:", valid_phone_numbers)
else:
    print("No valid phone numbers found.")
```
