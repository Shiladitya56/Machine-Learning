##Reading and displaying images using different libraries.
* comments have been explicitly included for readability.





```python
import cv2
# from google.colab.patches import cv2_imshow
import imageio
import matplotlib.pyplot as plt
from PIL import Image

# Read image using OpenCV
image_cv2 = cv2.imread('loveMath.jpg')

# Display image using OpenCV
plt.imshow(cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB))
plt.title('OpenCV Image')
plt.axis('off')
plt.show()

# Write image using OpenCV
cv2.imwrite('example_image_cv2.jpg', image_cv2)

# Read image using ImageIO
image_imageio = imageio.v3.imread('loveMath.jpg')

# Display image using ImageIO
plt.imshow(image_imageio)
plt.title('ImageIO Image')
plt.axis('off')
plt.show()

# Write image using ImageIO
imageio.imwrite('example_image_imageio.jpg', image_imageio)

# Read image using Matplotlib
image_matplotlib = plt.imread('loveMath.jpg')

# Display image using Matplotlib
plt.imshow(image_matplotlib)
plt.title('Matplotlib Image')
plt.axis('off')
plt.show()

# Write image using Matplotlib
plt.imsave('example_image_matplotlib.jpg', image_matplotlib)

# Read image using PIL
image_pil = Image.open('loveMath.jpg')

# Display image using PIL
image_pil.show()

# Write image using PIL
image_pil.save('example_image_pil.jpg')
```

##write the images using different libraries


*   cv2


```python
# Read the image
image = cv2.imread('loveMath.jpg')

# Convert and save as JPEG
cv2.imwrite('loveMath_cv2.jpg', image)

# Convert and save as PNG
cv2.imwrite('loveMath_cv2.png', image)
```

*   imageio


```python
# Read the image
image = imageio.imread('loveMath.jpg')

# Save as JPEG
imageio.imwrite('loveMath_imageio.jpg', image)

# Save as PNG
imageio.imwrite('loveMath_imageio.png', image)
```

* PIL

```python
# Open the image
image = Image.open('loveMath.jpg')

# Save as JPEG
image.save('loveMath_pil.jpg')

# Save as PNG
image.save('loveMath_pil.png')
```
