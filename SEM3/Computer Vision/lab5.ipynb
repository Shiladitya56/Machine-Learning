## Ideal low pass and Ideal high pass filter in frequency domain

### Importing the necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as pt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('dead.jpg',0)
cv2_imshow(img);
```

### Image in frequency domain

```python
F = np.fft.fft2(img)
pt.imshow(np.log1p(np.abs(F)),cmap='gray')
pt.axis('off');
```

### Shifiting low frequency component to centere of the image

```python
Fshift = np.fft.fftshift(F)
pt.imshow(np.log1p(np.abs(Fshift)),'gray')
pt.axis('off');
```

### Low pass filter

```python
M,N = img.shape
H = np.zeros((M,N), dtype=np.float32)
D0 = 50
for u in range(M):
  for v in range(N):
    D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
    if D <= D0:
      H[u,v] = 1
    else:
      H[u,v] = 0
pt.imshow(H,'gray')
pt.axis('off');
```

### Ideal Low pass filter

```python
Gshift = Fshift * H
pt.imshow(np.log1p(np.abs(Gshift)),'gray')
pt.axis('off');
```

### Inverse the shifted Fourier Transform

```python
G = np.fft.ifftshift(Gshift)
pt.imshow(np.log1p(np.abs(G)),'gray')
pt.axis('off');
```

### Inverse the fourier transformation

```python
g = np.abs(np.fft.ifft2(G))
pt.imshow(g,'gray')
pt.axis('off');
```

### High pass filter

```python
H = 1 - H
pt.imshow(H,'gray')
pt.axis('off');
```

### Ideal High Pass Filtering

```python
Gshift = Fshift * H
pt.imshow(np.log1p(np.abs(Gshift)),'gray')
pt.axis('off');
```

### Inverse the shifted Fourier Transform

```python
G = np.fft.ifftshift(Gshift)
pt.imshow(np.log1p(np.abs(G)),'gray')
pt.axis('off');
```

### Inverse Fourier Transform

```python
g = np.abs(np.fft.ifft2(G))
pt.imshow(g,'gray')
pt.axis('off');
```
