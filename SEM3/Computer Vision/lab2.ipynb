###Importing necessary libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

###Identity transformation

```python
image = cv2.imread('dog.jpeg')
# Identity transformation
identity_matrix = np.float32([[1, 0, 0],
                              [0, 1, 0]])
identity_transformed_image = cv2.warpAffine(image, identity_matrix, (image.shape[1], image.shape[0]))
cv2_imshow(identity_transformed_image)
```

###Reflect transformation

```python
image = cv2.imread('dog.jpeg')
# Reflect transformation along x-axis
reflect_matrix = np.float32([[-1, 0, image.shape[1]],
                             [0, 1, 0]])
reflect_transformed_image = cv2.warpAffine(image, reflect_matrix, (image.shape[1], image.shape[0]))
cv2_imshow(reflect_transformed_image)
```

###Scale transformation

```python
image = cv2.imread('dog.jpeg')
# Scale transformation
scale_factor_x = 1.5
scale_factor_y = 1.5
scale_matrix = np.float32([[scale_factor_x, 0, 0],
                           [0, scale_factor_y, 0]])
scaled_transformed_image = cv2.warpAffine(image, scale_matrix, (int(image.shape[1]*scale_factor_x), int(image.shape[0]*scale_factor_y)))
cv2_imshow(scaled_transformed_image)
```

###Rotate transformation

```python
image = cv2.imread('dog.jpeg')
# Rotate transformation
angle = 90
rotation_matrix = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), angle, 1)
rotated_transformed_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))
cv2_imshow(rotated_transformed_image)
```

###Shear transformation

```python
image = cv2.imread('dog.jpeg')
# Shear transformation
shear_matrix = np.float32([[1, 0.5, 0],
                           [0.5, 1, 0]])
sheared_transformed_image = cv2.warpAffine(image, shear_matrix, (int(image.shape[1]*1.5), int(image.shape[0]*1.5)))
cv2_imshow(sheared_transformed_image)
```
