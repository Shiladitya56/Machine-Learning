#We use the following image for our test purposes

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('dead.jpg')
cv2_imshow(image)
```

#Program 1

##Translation

```python
h, w = image.shape[:2]
tx = 100
ty = 50
trans = np.float32([[1, 0, tx], [0, 1, ty]])
trans_im = cv2.warpAffine(image, trans, (w, h))
cv2_imshow(trans_im)
```

##Scaling

```python
sx = 0.5
sy = 0.5
scale = np.float32([[sx, 0, 0], [0, sy, 0]])
scaled_im = cv2.warpAffine(image, scale, (w,h))
cv2_imshow(scaled_im)
```

##Rotation

```python
angle = 45
rotation = cv2.getRotationMatrix2D((w / 2, h / 2), angle, 1)
rotated_im = cv2.warpAffine(image, rotation, (w, h))
cv2_imshow(rotated_im)
```

#Program 2

##Average filter

```python
ks = (23, 23)
avg_im = cv2.blur(image, ks)
cv2_imshow(avg_im)
```

##Laplacian filter

```python
lap = np.array([[0, 1, 0],
                [1, -4, 1],
                [0, 1, 0]])
lap_im = cv2.filter2D(image, -1, lap)
cv2_imshow(lap_im)
```
