```python
import cv2
import numpy as np
import matplotlib.pyplot as pt
```

```python
img1=cv2.cvtColor(cv2.imread('1.jpeg'), cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(cv2.imread('2.jpeg'), cv2.COLOR_BGR2RGB)
img3=cv2.cvtColor(cv2.imread('3.jpg'), cv2.COLOR_BGR2RGB)
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("Original Image 1")
pt.imshow(img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("Original Image 2")
pt.imshow(img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("Original Image 3")
pt.imshow(img3)
pt.axis("off");
```

```python
point = np.array([[1, 1, 1],
                  [1, 6, 1],
                  [1, 1, 1]])
point_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=point)
point_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=point)
point_img3 = cv2.filter2D(src=img3, ddepth=-1, kernel=point)
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("Points Detected in Image 1")
pt.imshow(point_img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("Points Detected in Image 2")
pt.imshow(point_img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("Points Detected in Image 3")
pt.imshow(point_img3)
pt.axis("off");
```

```python
hor_line = np.array([[-1, -1, -1],
                     [2,   2,  2],
                     [-1, -1, -1]])
hor_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=hor_line)
hor_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=hor_line)
hor_line_img3 = cv2.filter2D(src=img3, ddepth=-1, kernel=hor_line)
ver_line = np.array([[-1, 2, -1],
                     [-1, 2, -1],
                     [-1, 2, -1]])
ver_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=ver_line)
ver_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=ver_line)
ver_line_img3 = cv2.filter2D(src=img3, ddepth=-1, kernel=ver_line)
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("Horizontal Line Detected in Image 1")
pt.imshow(hor_line_img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("Horizontal Line Detected in Image 2")
pt.imshow(hor_line_img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("Horizontal Line Detected in Image 3")
pt.imshow(hor_line_img3)
pt.axis("off");
```

```python
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("Vertical Line Detected in Image 1")
pt.imshow(ver_line_img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("Vertical Line Detected in Image 2")
pt.imshow(ver_line_img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("Vertical Line Detected in Image 3")
pt.imshow(ver_line_img3)
pt.axis("off");
```

```python
pos45_line = np.array([[-1, -1,  2],
                       [-1,  2, -1],
                       [ 2, -1, -1]])
pos45_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=pos45_line)
pos45_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=pos45_line)
pos45_line_img3 = cv2.filter2D(src=img3, ddepth=-1, kernel=pos45_line)
neg45_line = np.array([[ 2, -1, -1],
                       [-1,  2, -1],
                       [-1, -1,  2]])
neg45_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=neg45_line)
neg45_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=neg45_line)
neg45_line_img3 = cv2.filter2D(src=img3, ddepth=-1, kernel=neg45_line)
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("+45 Slanting in Image 1")
pt.imshow(pos45_line_img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("+45 Slanting in Image 2")
pt.imshow(pos45_line_img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("+45 Slanting in Image 3")
pt.imshow(pos45_line_img3)
pt.axis("off");
```

```python
pt.figure(figsize=(20,20))
pt.subplot(1,4,1)
pt.title("-45 Slanting in Image 1")
pt.imshow(neg45_line_img1)
pt.axis("off")
pt.subplot(1,4,2)
pt.title("-45 Slanting in Image 2")
pt.imshow(neg45_line_img2)
pt.axis("off")
pt.subplot(1,4,3)
pt.title("-45 Slanting in Image 3")
pt.imshow(neg45_line_img3)
pt.axis("off");
```
