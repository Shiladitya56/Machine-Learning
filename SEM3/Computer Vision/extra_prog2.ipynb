```python
import urllib
import cv2 as cv2
import numpy as np
from google.colab.patches import cv2_imshow
from scipy import ndimage
```

```python
url = 'https://hips.hearstapps.com/hmg-prod/images/closeup-of-a-black-russian-tsvetnaya-bolonka-royalty-free-image-1681161235.jpg?crop=0.563xw:1.00xh;0.204xw,0&resize=1200:*'
url_response = urllib.request.urlopen(url)
img_array = np.array(bytearray(url_response.read()), dtype=np.uint8)
img = cv2.imdecode(img_array, -1)
img = cv2.resize(img, (400,400))
cv2_imshow(img)
```

###define the kernels

```python
identity = np.array([[0,0,0],
                     [0,1,0],
                     [0,0,0]])
edge = np.array([[-1,-1,-1],
                 [-1,6,-1],
                 [-1,-1,-1]])
sharp = np.array([[0,-1,0],
                  [-1,6,-1],
                  [0,-1,0]])
box = np.ones((3,3),np.float32)/6
gauss = np.array([[1,2,1],
                  [2,4,2],
                  [1,2,1]])/16
```

##Identity

```python
cv2.filter2D(src=img, ddepth=-1, kernel=identity)
```

##Edge Detection

```python
cv2.filter2D(src=img, ddepth=-1, kernel=edge)
```

##Sharpen

```python
cv2.filter2D(src=img, ddepth=-1, kernel=sharp)
```

##Box Blur

```python
cv2.filter2D(src=img, ddepth=-1, kernel=box)
```

##Gaussian Blur

```python
cv2.filter2D(src=img, ddepth=-1, kernel=gauss)
```

##Trying out the different modes of convolve

```python
for i in ['reflect', 'constant', 'nearest', 'mirror', 'wrap']:
  print('using mode ',i)
  cv2_imshow(ndimage.convolve(img[:,:,0], identity, mode=i))
```
