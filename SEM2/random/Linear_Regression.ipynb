```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/ipynb/datasets/computers.csv')
df
```

```python
df.describe()
```

```python
pt.figure(figsize=(4,3))
pt.xlabel('Units')
pt.ylabel('Minutes')
pt.tight_layout()
pt.axhline(int(np.mean(df.Minutes)), c='r')
pt.annotate('Average Repair Time', xy=(1,102))
pt.scatter(df['Units'], df['Minutes'], color='orchid', marker='x');
```

```python
x = df.Units
y = df.Minutes
n = float(len(x))
l = 0.0001
m = 0
c = 0
```

```python
def lin_reg():
  global m,c
  for i in range(1000):
    yp = m*x + c
    dm = (-2 / n) * sum(x * (y - yp))
    dc = (-2 / n) * sum(y - yp)
    m = m - l * dm
    c = c - l * dc
def pred(x):
  return m*x + c
```

```python
lin_reg()
```

```python
y_pred = []
for i in x:
  y_pred.append(pred(i))
```

```python
pt.xlabel('Units')
pt.ylabel('Minutes')
pt.tight_layout()
pt.plot(x, y_pred, color='lightgreen')
pt.scatter(df['Units'], df['Minutes'], color='red', marker='x');
```

```python
y_step = []
m = 0
c = 0
def lin_reg_step():
  for i in range(1000):
    global m,c
    yp = m*x + c
    dm = (-2 / n) * sum(x * (y - yp))
    dc = (-2 / n) * sum(y - yp)
    m = m - l * dm
    c = c - l * dc
    k = -1
    temp = []
    for j in x:
      temp.append(m*j+c)
    y_step.append(temp)
lin_reg_step()
```

```python
for i in range(0,1000,10):
  pt.plot(x, y_step[i])
pt.xlabel('Units')
pt.ylabel('Minutes')
pt.tight_layout()
pt.scatter(df['Units'], df['Minutes'], color='red', marker='o', zorder=999);
```
